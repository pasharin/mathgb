## Задача №4

```python
x = np.linspace(-10, 10, 100) 
k = 1
k2 = int(input('Введите значение k: '))          
y1 = np.cos(k * x)                      
y2 = np.cos(k2 * x)  

fig, ax = plt.subplots()                        # будет 1 график, на нем:
ax.plot(x, y1, color="green", label="y(x)")      # функция y1(x), синий, надпись y(x)
ax.plot(x, y2, color="orange", label="y'(x)")      # функция y2(x), красный, надпись y'(x)
ax.set_xlabel("x")                              # подпись у горизонтальной оси х
ax.set_ylabel("y")                              # подпись у вертикальной оси y
ax.legend()                                     # показывать условные обозначения

plt.show()                                      # показать рисунок
```
